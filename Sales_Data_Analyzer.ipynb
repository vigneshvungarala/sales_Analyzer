## Loading the data

```python
import pandas as pd
sales_data=pd.read_csv("raw_sales_data.csv")
sales_data.head()
```

```python
sales_data.info()
```

## Data Cleaning

```python
sales_data['Date']=pd.to_datetime(sales_data['Date'],errors='coerce')
sales_data.head()
```

```python
sales_data['Date'] = sales_data['Date'].fillna(method='ffill')
sales_data.info()
```

```python
sales_data.isna().sum() #check for missing values
```

```python
# fill missing values with column mean
sales_data['Total Sales'].fillna(sales_data['Total Sales'].mean(), inplace=True)
sales_data['Customer Age'].fillna(sales_data['Customer Age'].mean(), inplace=True)
# Check for missing values after filling
print(sales_data.isnull().sum())
```

```python
sales_data.drop_duplicates(inplace=True) #removing duplicates
```
